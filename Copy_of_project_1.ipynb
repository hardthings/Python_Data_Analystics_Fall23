```python
import pandas as pd
import numpy as  np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')

```

```python
# Overview of the entire dataset
df.describe()
```

```python
# Data for Alg3/Trig Regents Exam for bayside and Fancis Lewis High Schoolin 2016
df_BaysideFrancis= df[(df['School Name'].isin(['Francis Lewis High School','Bayside High School']) &
                       (df['Regents Exam']=='Algebra2/Trigonometry') &
                       (df['Year']==2016))]
```

```python
df_BaysideFrancis
```

```python
#  only columns for Alg2/Trigonometry mean scores for both schools in 2016
df_Exam2016=df_BaysideFrancis[['School Name','Mean Score']]
df_Exam2016
```

```python
df_Exam2016['Mean Score']= pd.to_numeric(df_Exam2016['Mean Score'], errors='coerce')
```

```python
#Statistics for the mean score column
df_Exam2016['Mean Score'].describe()
```

```python
#Average of all the mean scores for Bayside High School
df_meanscore_bayside= df_Exam2016.loc[df_Exam2016['School Name']== 'Bayside High School','Mean Score'].mean()
df_meanscore_bayside
```

```python
#Average of all the mean scores for Francis Lewis High School
df_meanscore_francislewis= df_Exam2016.loc[df_Exam2016['School Name']=='Francis Lewis High School','Mean Score'].mean()
df_meanscore_francislewis
```

```python
df_meanscore_by_school = df_Exam2016.groupby('School Name')['Mean Score'].mean()

df_meanscore_by_school.plot(kind='bar', color=['red','blue'])
plt.title('Mean Score for Algebra 2/Trigonometry Regents Exam 2016')
plt.xlabel('School')
plt.ylabel('Mean Score')
```

CONLCUSION: Based on the analysis, we can see that Francis Lewis High School had a higher average mean score value of 74.8 and Bayside Highe School had a value of 60 for the Algebra2/Trigonomerty Regents Exam. the lowest mean score on the regents Exasm wa 46.6 and the highest mean score wa 82.8.
